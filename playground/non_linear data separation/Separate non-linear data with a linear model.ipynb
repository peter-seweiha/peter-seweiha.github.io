## How can we use a linear model to separate data that's not linearly separable

As part of my interest in Linear Models, I was playing to prove that I can feature engineer the data so it's linearly separable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data_shape.csv')
```

```python
# 0/1 can't be separated with a line

plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sns.scatterplot( x= data['x'], y=data['y'], hue = data[ 'label'], legend=False)
plt.title('Plotting the original data')

# with feature engineering, now we can separate them with a linear model e.g. Logistic regression
plt.subplot(1,2,2)
sns.scatterplot( x= (data['x']-6)**6, y=(data['y']-5)**4, hue = data[ 'label'], legend=False)
plt.title('Plotting same data after Feature Engineering')
plt.show()
```
