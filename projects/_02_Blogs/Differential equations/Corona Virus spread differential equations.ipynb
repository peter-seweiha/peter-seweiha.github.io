```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
susceptable = 0.99
infected = 0.01
removed = 0
max_time = 6000
delta_t = 0.01
```

```python
infection_rate = 0.6
recovery_rate = 0.23
```

```python
sus_change = []
inf_change = []
rem_change = []

sus_change.append(susceptable)
inf_change.append(infected)
rem_change.append(removed)
```

```python
for day in range(0,max_time ):

    Srate = - infection_rate * susceptable * infected
    Irate = infection_rate * susceptable * infected - (recovery_rate* infected)
    Rrate = recovery_rate* infected
    
    susceptable = susceptable + (Srate * delta_t) 
    infected = infected + (Irate * delta_t)
    removed = removed + (Rrate * delta_t)
    
    sus_change.append(susceptable)
    inf_change.append(infected)
    rem_change.append(removed)
```

```python
model = pd.DataFrame({'Susceptable':sus_change, 'Infected': inf_change, 'Recovered':rem_change  })
```

```python
model.plot(figsize=(14,8))
```
