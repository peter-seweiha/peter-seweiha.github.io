# Forecast the daily number of restaurant customers

**Dataset:**
Restaurant Visitors dataset that was inspired by a recent Kaggle competition. The data considers daily visitors to four restaurants located in the United States, subject to American holidays. For the exogenous variable we'll see how holidays affect patronage. The dataset contains 478 days of restaurant data, plus an additional 39 days of holiday data for forecasting purposes.

**Problem statement:** we have the daily number of customers for a restaurant chain (from Jan-2016 to Apr-2017, 478 observations) - we need to find the most accurate method to predict 4 weeks ahead given that the data exhibit:

- Trends
- Seasonality 
- Exogenous factors (holidays)

**Approach:**

1. A forecast will be conducted and assessed using SARIMAX
2. Same will be accomplished using fbprophit and a comparison
3. See how far can the results be replicated using traditional regression models


```python
# Import key liberaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Understanding and prepare the time series for modelling
from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from statsmodels.tsa.stattools import adfuller               # to run stationarity test
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf # for determining (p,q) orders
from sklearn.preprocessing import LabelEncoder               # to encode weekday for regression


# Load forecasting tools
from pmdarima import auto_arima                              # for determining ARIMA orders
from statsmodels.tsa.statespace.sarimax import SARIMAX

from xgboost import XGBRegressor


# Model Evaluation
from statsmodels.tools.eval_measures import mse,rmse

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset
data = pd.read_csv('RestaurantVisitors.csv', index_col='date', parse_dates=True)
data.head()
```

```python
forecast_period = 28  # in days
```

```python
# I'll clean the data by taking out missing data and consider only total customers
data = data.dropna()
data = data.drop(['rest1', 'rest2', 'rest3', 'rest4'], axis=1)

# Convert 'total' column from float to int
data['total'] = data['total'].astype(int)

# Set the reight datetime frequency
data.index.freq = 'D'
```

```python
data.head()
```

```python
# plot the timeseries
ax = data['total'].plot(figsize=(16,5))
plt.ylabel('no of customers')

# Add horizontal lines to represent holidays
for x in data[data['holiday']==1].index:       
    ax.axvline(x=x, color='red', alpha = 0.4);  
```

# Model & forecast using SARIMAX

1. Perform ETS Decomposition
2. Test for Stationarity
3. plot ACF & PACF 
4. Obtain ARIMA recommended orders (using gridsearch: pmdarima.auto_arima ) 
5. Split the data for modelling
6. Fit ARIMA model
7. Assess the model
8. Add the Exogenous variable (Public Holidays)
9. Re-assess the model

### 1. Perform ETS Decomposition

```python
# Perform ETS Decomposition
decom = seasonal_decompose(data['total'] )
decom.plot();
```

```python
# Plotting the trend line to view visually the potential influence of holidays
ax = decom.trend.plot(figsize=(16,5))
plt.ylabel('Actual no of customers')

# Add horizontal lines to represent holidays
for x in data[data['holiday']==1].index:       
    ax.axvline(x=x, color='red', alpha = 0.4)
```

### 2. Test for stationarity
**Using "Augmented DickeyFuller test"**


```python
# A copied code (from www.pieriandata.com) to enhance the "Augmented DickeyFuller test" results readability

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
# Run Augmented Dickey-Fuller Test
adf_test(data['total'])
```

### 3. Plot ACF & PACF graphs
Auto-Correlation and Partial Auto-Correlation Functions

Plotting the magnitude of the autocorrelations over the first few (20-40) lags can say a lot about a time series.

```python
# Plot ACF & PACF graphs
plot_acf(data['total'], lags=40);
plot_pacf(data['total'], lags=40);
```

### 4. Obtain ARIMA recommended orders 

```python
# Determine the recommended ARIMA order using gridsearch (pmdarima.auto_arima)
auto_arima( data['total'], seasonal=True, m =7, random_state = 12, n_jobs=-1).summary()
```

This provides an ARIMA Order of (1,0,0) and a seasonal order of (2,0,0,7)

### 5. Split the data

```python
# Split dataset into training & testing
train_data = data[: -forecast_period ]
test_data = data[ -forecast_period :]
```

### 6. Fit ARIMA model (SARIMA)
NOTE: To avoid a <tt>ValueError: non-invertible starting MA parameters found</tt> we're going to set <tt>enforce_invertibility</tt> to False.

```python
# Fit an ARIMA model
model = SARIMAX(train_data['total'],order=(1,0,0),seasonal_order=(2,0,0,7),enforce_invertibility=False)
results = model.fit()
results.summary()
```

### 7. Assess the model

```python
# Obtain predicted values
start=len(train_data)
end=len(train_data)+len(test_data)-1
y_pred = results.predict(start=start, end=end, dynamic=False).rename('SARIMA(1,0,0)(2,0,0,7) Predictions')

# plot model performance
ax = test_data['total'].rename('Actual no of customers').plot(figsize = (12,6), legend=True)
y_pred.plot(legend=True, color ='green', ls='--')

# print accuracy measurements
print(f"RMSE: {np.round(rmse(test_data['total'], y_pred),2)}")
print(f"MSE: {np.round(mse(test_data['total'], y_pred),2)}")

# Add horizontal line to represent holidays
for x in test_data[test_data['holiday']==1].index:       
    ax.axvline(x=x, color='red', alpha = 0.4)
```

### 8. Add the Exogenous variable 

```python
# Add the Exogenous variable (the public holidays)
model = SARIMAX(train_data['total'],order=(1,0,0),seasonal_order=(2,0,0,7),
                exog=train_data['holiday'], enforce_invertibility=False)
results = model.fit()
results.summary()
```

### 9. Re-assess model performance

```python
# Obtain predicted values
start=len(train_data)
end=len(train_data)+len(test_data)-1
y_pred_ar = results.predict(start=start, end=end, exog=test_data[['holiday']],
                         dynamic=False).rename('SARIMAX(1,0,0)(2,0,0,7) Predictions')

# print accuracy measurements
print(f"RMSE: {np.round(rmse(test_data['total'], y_pred_ar),2)}")
print(f"MSE: {np.round(mse(test_data['total'], y_pred_ar),2)}")

# plot model performance
ax = test_data['total'].rename('Actual no of customers').plot(figsize = (12,6), legend=True)
y_pred_ar.plot(legend=True, color ='green', ls='--')

# Add horizontal line to represent holidays
for x in test_data[test_data['holiday']==1].index:       
    ax.axvline(x=x, color='red', alpha = 0.4)
```

# Forecast using FaceBook Prophet

This package have some special requirements, however the output is great in terms of clarity, readability and the actual performance is great too!

- Data should be stored in DataFrame
- Date column is named 'ds' and formatted as datetime
- The time series is stored in a column named 'y'
- Holidays are stored in a column named 'holiday' in a str format

```python
# Load FaceBook Prophet package
from fbprophet import Prophet
```

```python
# Load and prepare the dataset for fbprophet
data = pd.read_csv('RestaurantVisitors.csv', index_col='date', parse_dates=True)

# Prepare the dataset
data = data.dropna()
data['ds']= data.index
data['y'] = data['total']
data = data.drop(['weekday', 'total', 'rest1', 'rest2', 'rest3', 'rest4'], axis=1)
data.rename(columns={'holiday_name':'holiday', 'holiday':'holiday_1'}, inplace=True)
```

```python
# Split data
train_data = data[: -forecast_period ]
test_data = data[ -forecast_period :]
```

```python
m = Prophet(holidays=data[['holiday', 'ds']])
m.fit(train_data[['holiday','ds','y']])

#Create "future" placeholder dataframe
future = m.make_future_dataframe(periods=forecast_period ,freq = 'D')

forecast = m.predict(future)
```

```python
# Converting the forecast only to a series
fd_pred = forecast[['ds','yhat']]
fd_pred.index = fd_pred['ds']
y_pred_fb = fd_pred.drop('ds', axis=1)[-forecast_period :]
y_pred_fb = y_pred_fb['yhat'].rename('fbprophet Predictions')
```

```python
# print accuracy measurements
print(f"RMSE: {np.round(rmse(test_data['y'], y_pred_fb),2)}")
print(f"MSE: {np.round(mse(test_data['y'], y_pred_fb),2)}")
      
# plot model performance
pd.plotting.register_matplotlib_converters()    # this line is needed to aviod a clash between fbprophet & pandas
      
# plot model performance
ax = test_data['y'].rename('Actual no of customers').plot(figsize = (12,6), legend=True)
y_pred_fb.plot(legend=True, color ='green', ls='--')

# Add horizontal line to represent holidays
for x in test_data[test_data['holiday_1']==1].index:       
    ax.axvline(x=x, color='red', alpha = 0.4)
```

## Time Series Forecast using traditional regression (XGB)

```python
# Import and clean the same dataset
data = pd.read_csv('RestaurantVisitors.csv', index_col='date', parse_dates=True)

# clean & prepare data
data = data.dropna()
data = data.drop([ 'holiday_name', 'rest1', 'rest2', 'rest3',
       'rest4'], axis=1)
data['total']= data['total'].astype(int)
```

```python
# Change the data lay-out to suit a traditional regression model

d_1 = []
d_2 = []
d_3 = []
d_4 = []
d_5 = []
d_6 = []
d_7 = []

for i in range(0, 471):
    d_1.append(data['total'][i])
    d_2.append(data['total'][i+1])
    d_3.append(data['total'][i+2])
    d_4.append(data['total'][i+3])
    d_5.append(data['total'][i+4])
    d_6.append(data['total'][i+5])
    d_7.append(data['total'][i+6])

data = data[7:]

data['d_1']= d_1
data['d_2']= d_2
data['d_3']= d_3
data['d_4']= d_4
data['d_5']= d_5
data['d_6']= d_6
data['d_7']= d_7
```

```python
# Encode the weekday feature
le = LabelEncoder()
data['weekday'] = le.fit_transform(data['weekday'])
```

```python
# Split the dataset into training & testing
X_train = data.drop(['total'], axis=1)[:-forecast_period]
X_test = data.drop(['total'], axis=1)[-forecast_period:]
y_train = data['total'][:-forecast_period]
y_test = data['total'][-forecast_period:]
```

```python
# Fit as XGBoost regression model
reg = XGBRegressor()
reg.fit(X_train,y_train)
y_pred_xgb = reg.predict(X_test)
```

```python
predictions = pd.DataFrame(y_test)
predictions['yhat']= y_pred_xgb

# print accuracy measurements
print(f"RMSE: {np.round(rmse(y_test, y_pred_xgb),2)}")
print(f"MSE: {np.round(mse(y_test, y_pred_xgb),2)}")

# Plot predictions against actual data
ax = predictions['total'].rename('Actual no of customers').plot(figsize=(12,6), legend=True)
predictions['yhat'].rename('XGB Predictions').plot(color='green', ls='--', legend=True)
for x in X_test[X_test['holiday']==1].index:
    ax.axvline(x=x, color='red', alpha = 0.4)
```

## The winning Algorithm is ...!

```python
print(f"FaceBook prophet's error (RMSE): {np.round(rmse(y_test, y_pred_fb),2)}")
print(f"SARIMAX's error (RMSE): {np.round(rmse(y_test, y_pred_ar),2)}")
print(f"XGBoost's error (RMSE): {np.round(rmse(y_test, y_pred_xgb),2)}")

# Plot predictions against actual data
ax = predictions['total'].rename('Actual no of customers').plot(figsize=(12,6), legend=True, color='black')
predictions['yhat'].rename('XGB Predictions').plot(color='green', ls='--', legend=True, alpha=0.8)
y_pred_ar.rename('SARIMAX Predictions').plot(color='blue', ls='--', legend=True, alpha=0.8)
y_pred_fb.rename('FaceBook prophet Predictions').plot(color='red', ls='--', legend=True, alpha=0.8)


for x in X_test[X_test['holiday']==1].index:
    ax.axvline(x=x, color='grey', alpha = 0.8)
```

### Learnings:

- FaceBook Prophet provided the highest accuracy for this type of problem and it proved to be intuitive and easy to use
- ARIMA based model (SARIMAX) is the second best (very close to XGB though). It requires time to verify the assumptions however doing so you learn a lot about the timeseries
- XGBoost's performance was similar to ARIMA (for my surprise!). I assumed it will not be able to take advantage of the intrecacies of a time series but it did. It just required some time to put the data in the right shape and some understanding about the data (e.g. how many features you'll break the time series into?)
