## Organising the dataset for training

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import date
from IPython.display import Markdown, display
```

```python
filename = ('SampleData.csv')
role_lookup = ('') 
```

```python
data = pd.read_csv(filename)
```

```python
# Remove uneeded columns from dataset
data = data[['CaseId', 'ActivityName', 'StartTimestamp']]

# Convert CaseId column to string
data['CaseId'] = data['CaseId'].astype('str')
```

```python
unique_cases = data['CaseId'].drop_duplicates().tolist()
```

```python
rows = []

# Activities add before and after should be same size
add_before_list = ['A', 'A', 'A', 'A', 'A']
add_after_list = ['Z', 'Z', 'Z', 'Z', 'Z']

for case in unique_cases:
    
    case_activities = data[data['CaseId']==case]
    
    # Sort activities based on time stamp
    case_activities = case_activities.sort_values(by='StartTimestamp', ascending= True)
    
    # Convert activities to list
    case_activities = case_activities['ActivityName'].tolist()
    
    # Add dummy activities before and after to help prep for model training
    case_activities_training = add_before_list + case_activities + add_after_list
    


    for i in range (len(case_activities_training)-len(add_before_list)):
        
        # Capture a series of 5 sequential activities at a time and link to Case Id
        sequence = case_activities_training[i:i+(len(add_before_list)+1)]
        sequence = [case] + sequence
        rows.append(sequence)
```
