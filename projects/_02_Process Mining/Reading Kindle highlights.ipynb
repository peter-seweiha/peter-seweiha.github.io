```python
import pandas as pd


# with open(filename, encoding = 'cp850') as csv_file:
#     csv_reader = csv.reader(csv_file, delimiter=',')


# Function to split a text file by a delimiter and append to a DataFrame
def split_text_to_dataframe(input_filename):
    segments = []
    current_segment = []

    # Open the input text file for reading
    with open(input_filename, 'r',  encoding = 'utf-8') as input_file:
        for line in input_file:
            # Check if the line contains the delimiter "========"
            if "========" in line:
                if current_segment:
                    segments.append(current_segment)
                current_segment = []
            else:
                # Add the line to the current segment
                current_segment.append(line.strip())

        # Append the last segment if it exists
        if current_segment:
            segments.append(current_segment)

    # Create a DataFrame from the segments
    df = pd.DataFrame(segments)

    return df

# Usage example
input_filename = "C:/Users/peter/OneDrive/PETER/Desktop/My Clippings.txt"  # Replace with your input file name
result_dataframe = split_text_to_dataframe(input_filename)
```

```python
# Define the index of the column from which you want to start concatenating
start_column_index = 3  # Index 2 corresponds to the third column

# Concatenate all text from the specified column index onwards
def concatenate_columns_from_index_onwards(row):
    return ' '.join([str(value) if value is not None else '' for value in row[start_column_index:]])

# Create a new column 'Concatenated' by applying the function to each row
result_dataframe['Concatenated'] = result_dataframe.apply(concatenate_columns_from_index_onwards, axis=1)
```

```python
data = pd.DataFrame({'title':result_dataframe[0], 'pages & date':result_dataframe[1], 'highlights & notes':result_dataframe['Concatenated']  })
```

```python
# Define a function to extract the desired parts
import re
def extract_parts(text):
    match = re.match(r'- Your (\w+) on (page \d+) \| .+? (\d{1,2} \w+ \d{4})', text)
    if match:
        return match.group(1), match.group(2), match.group(3)
    else:
        return None, None, None

# Apply the function to the DataFrame and create new columns
data[['Action', 'Page', 'Date']] = data['pages & date'].apply(extract_parts).apply(pd.Series)

```

```python
data = data.drop('pages & date' , axis=1)
```
